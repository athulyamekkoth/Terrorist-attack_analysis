```python
t=pd.read_csv("terror_gtd.csv",encoding='latin1')
```

```python
# Q1)How many attacks happened in India?

t[t['country_txt']=='India'].shape[0]
```

```python
# Q2) How many attacks happened in India and upto 3 people were killed?

t1=t[(t['country_txt']=='India')&(t['nkill']<=3)]
t1.shape[0]
```

```python
# Q3) Extract the city and summary for attacks above

t1[['city','summary']]
```

```python
# Q4) In a single terror incident in India, find out top 5 cities by number killed

t2=t[t['country_txt']=='India']
t3=t2.groupby('city')['nkill'].sum()
t4=t3.sort_values(ascending=False).head(5)
t4
```

```python
# Q5) In a single terror incident in India, find out top 5 cities by number killed and wounded

t2= t[t['country_txt'] == 'India']
t5 =t2.groupby('city')[['nkill', 'nwound']].sum()
t5['total'] = t5['nkill'] +t5['nwound']
total = t5['total'].sort_values(ascending=False).head(5)
total
```

```python
# Q6) How many attacks were successful that were suicide attacks?

t[(t['success']==1)&(t['suicide']==1)].shape[0]
```

```python
# Q7)  Label all the incidents where the number killed was more than 5 as severe.

t6=t['incidents']=t['nkill'].apply(lambda x:'severe'  if x>5 else 'not severe')

#displaying the severe incidents

t[t['incidents']=='severe']
```

```python
# Q8) write a function to label an incident that was both successful and suicidal

def incident(success,suicide):
    if success==1 and suicide==1:
        return 'successful suicidal attack'
    else:
        return 'Unsuccessful suicidal attack'
        
```

```python
t['incident_1']=t.apply(lambda x:incident(x['success'],x['suicide']),axis=1)
t.head(1)
```

```python
# Q9)

def category(c):
    if c=='Afghanistan' or c=='Pakistan' or c=='India':
        return "level 1 category"
    else:
        return "ROW"
```

```python
t['Local']=t['country_txt'].apply(category)

#displaying first 4 rows of t['Local']

t[['country_txt','Local']].head(4)
```
